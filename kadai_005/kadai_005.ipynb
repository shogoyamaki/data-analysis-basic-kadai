```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, tight_layout=True)

# 折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)

# sin関数
x1 = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x1)
axes[0, 1].plot(x1, y1)

# ヒストグラム
data = np.random.randn(1000)
axes[0, 2].hist(data, bins=30)

# 散布図
x2 = np.random.uniform(-5, 2, 100)
y2 = np.random.uniform(-5, 2, 100)
axes[1, 0].scatter(x2, y2)

plt.show()
```
